```python
import gymnasium as gym
import numpy as np
import gym_gridworlds

# Initialize the custom map with BEV observations.
env = gym.make(
    "Gym-Gridworlds/NineRooms-Loop-21x21-v0",
    observation_type="bev",
)

# Store the initial observation and scalar state id for inspection.
initial_observation, initial_info = env.reset(seed=0)
initial_state = initial_info["state"]
initial_agent_pos = env.unwrapped.agent_pos

# Take a few deterministic moves: right, right, down, down, left.
# Action ids in this environment are: left=0, right=1, down=2, up=3, stay=4.
actions = [1, 1, 2, 2, 0]
trajectory = []
for action in actions:
    observation, reward, terminated, truncated, info = env.step(action)
    trajectory.append(
        {
            "action": action,
            "state": info["state"],
            "agent_pos": env.unwrapped.agent_pos,
            "reward": reward,
            "terminated": terminated,
            "truncated": truncated,
        }
    )
    if terminated or truncated:
        break

# Store one later observation for comparison with the initial one.
moved_observation = observation
moved_state = info["state"]
moved_agent_pos = env.unwrapped.agent_pos

# Useful quick checks. Channel order: unseen, empty, wall, left-door, right-door, down-door, up-door, agent.
print("initial shape:", initial_observation.shape)
print("moved shape:", moved_observation.shape)
print("initial state / pos:", initial_state, initial_agent_pos)
print("moved state / pos:", moved_state, moved_agent_pos)
print("trajectory:", trajectory)
print("initial visible cells:", int((initial_observation[..., 0] == 0).sum()))
print("moved visible cells:", int((moved_observation[..., 0] == 0).sum()))
```
